```python
A.drop(['Time','Sno','ConfirmedIndianNational','ConfirmedForeignNational'],inplace =True, axis = 1)
A.head()
```

```python
A.tail()
```

```python
A['ActiveCases'] = A['Confirmed'] - (A['Cured'] + A['Deaths'])
A.tail()
```

```python
top_10_active_cases = A.groupby(by = 'State/UnionTerritory').max()[['ActiveCases' , 'Date']].sort_values(by='ActiveCases',ascending = False).reset_index()
```

```python
top_10_active_cases
```

```python
fig = plt.figure(figsize=(18,8))
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y='ActiveCases', x='State/UnionTerritory')
plt.title('Top 10 states with most active cases',size = 25)
plt.xlabel('State')
plt.ylabel('Active Cases')
```

```python
ax = sns.barplot(data = top_10_active_cases.iloc[:10], y='ActiveCases', x='State/UnionTerritory')
plt.title('Top 10 states with most active cases',size = 25)
plt.xlabel('State')
plt.ylabel('Active Cases')


fig = plt.figure(figsize=(12,6))
ax = sns.lineplot(data = A[A['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],x='Date',y='ActiveCases',hue = 'State/UnionTerritory')

ax.set_title("Top 5 Affected States in India",size = 16)
```

```python
statewise = pd.pivot_table(A , values=['Cured','Deaths','Confirmed'],index = 'State/UnionTerritory',aggfunc='max')
statewise['Recovery Rate'] = statewise['Cured']*100 /  statewise['Confirmed']
statewise['Death Rate'] = statewise['Deaths']*100 /  statewise['Confirmed']
statewise.style.background_gradient()
```
