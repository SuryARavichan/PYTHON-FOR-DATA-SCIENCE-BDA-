```python
male = vaccine['Male(Individuals Vaccinated)'].sum()
female = vaccine['Female(Individuals Vaccinated)'].sum()
tg=vaccine['Transgender(Individuals Vaccinated)'].sum()
labels = ['Male','Female','tg']
x = np.array([male,female,tg])
plt.pie(x,labels = labels,autopct='%.1f%%')
plt.title('vaccination distribution across Gender')
plt.legend()
plt.show()
```

```python
young_middle_age = vaccine['18-44 Years(Individuals Vaccinated)'].sum()
upper_middle_age = vaccine['45-60 Years(Individuals Vaccinated)'].sum()
senior_citizens=vaccine['60+ Years(Individuals Vaccinated)'].sum()
labels = ['Y&M','U_MA','SCitizens']
x = np.array([young_middle_age,upper_middle_age,senior_citizens])
fig = plt.figure(figsize=(12,10))
plt.pie(x,labels = labels,autopct='%.4f%%')
plt.title('vaccination distribution across Age groups')
plt.legend()
plt.show()
```
