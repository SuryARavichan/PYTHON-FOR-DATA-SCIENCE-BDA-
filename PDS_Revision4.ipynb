```python
np.random.seed(0)
df6 = pd.DataFrame(np.random.randn(7, 3))
df6
```

```python
#11 Drop the row containing two or more null values for the above DF
```

```python
df6.dropna(thresh=3)  #thresh is the count for non-NA
```

```python
#12. For the df6 DF fill the missing value with '0'
```

```python
df6.fillna(0)
```

```python
#13 For the df6 DF fill missing values with mean values
```

```python
df6.fillna(df6.mean())
```

```python
#14 for the df6 DF fill column '1' with mean and column '2' with median
```

```python
df6.fillna({1:df6[1].mean(),2:df6[2].median()})
```
